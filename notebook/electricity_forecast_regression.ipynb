```python
# test train split

# from sklearn.model_selection import train_test_split
# x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=7)

def get_train_split(df, target_col, size=0.7):

    y = df[target_col]
    x = df.drop(columns=[target_col])

    test_index = int(len(x) * size)
    x_train = x[1:test_index]
    x_test = x[test_index:]
    y_train = y[1:test_index]
    y_test = y[test_index:]
    return x_train, x_test, y_train, y_test
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def get_metrics(y_pred, y_test):
    mape = np.mean(np.absolute((y_test - y_pred)/y_test)) * 100
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    r2 = r2_score(y_test, y_pred)

    print(f"MAPE: {mape:.2f}%")
    print(f"MAE: {mae:.2f}")
    print(f"MSE: {mse:.2f}")
    print(f"RMSE: {rmse:.2f}")
    print(f"R2: {r2:.2f}")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn import set_config

set_config(display="diagram")

def make_fit_pipeline(x_train, y_train, estimator):

    num_cols = x_train.select_dtypes("float64").columns
    
    num_pipeline = Pipeline(steps=[
        ("imputer", SimpleImputer(strategy="mean")),                      
        ("scaler", StandardScaler())
    ])

    col_trans = ColumnTransformer(transformers=[
        ("num_scaler", num_pipeline, num_cols)
    ],
    remainder="drop")

    pipeline = Pipeline(steps=[
        ("scale", col_trans),
        ("estimator", estimator)
    ])
    
    pipeline.fit(x_train, y_train)
    return pipeline

def plot_predictions(y_pred, y_test):
    df = pd.DataFrame({"actual":y_test})
    df["prediction"] = y_pred
    df.set_index(y_test.index, inplace=True)
    df.plot(figsize=(12,4))

```

```python
x_train, x_test, y_train, y_test  = get_train_split(raw_df, "nat_demand")
```

```python
# baseline of prediction 1 hr ahead
y_pred = y_test.shift(1).bfill()
plot_predictions(y_pred, y_test)
get_metrics(y_pred, y_test)
```

```python
pipeline = make_fit_pipeline(x_train, y_train, LinearRegression())
y_pred = pipeline.predict(x_test)
plot_predictions(y_pred, y_test)
get_metrics(y_pred, y_test)
```

```python
from sklearn.linear_model import ElasticNet
pipeline = make_fit_pipeline(x_train, y_train, ElasticNet(l1_ratio=0.5))
y_pred = pipeline.predict(x_test)
plot_predictions(y_pred, y_test)
get_metrics(y_pred, y_test)
```

```python
# add last day's hr lags 
def add_lags(df, col_name, lags = 12):
    lag_df = df.copy(deep=True)
    for i in range(lags):
        lag_df["lag" + str(i+1)] = lag_df[col_name].shift(i+1)
    return lag_df.dropna()

raw_df_w_lags = add_lags(raw_df, "nat_demand", 24)

x_train, x_test, y_train, y_test = get_train_split(raw_df_w_lags, "nat_demand")

pipeline = make_fit_pipeline(x_train, y_train, ElasticNet(l1_ratio=0.5))
y_pred = pipeline.predict(x_test)
plot_predictions(y_pred, y_test)
get_metrics(y_pred, y_test)
```

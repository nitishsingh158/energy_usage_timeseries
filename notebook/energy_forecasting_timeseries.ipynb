```python
import pandas as pd
import seaborn as sns

sns.set_style("darkgrid")
```

```python
# https://www.kaggle.com/code/alcheng10/forecasting-electricity-pjm-us-east-coast-grid
```

```python
filepath = "./dataset/PJME_hourly.csv"
raw_data = pd.read_csv(filepath_or_buffer=filepath, parse_dates=["Datetime"])

raw_data = raw_data.set_index("Datetime")
raw_data = raw_data.loc[raw_data.index.drop_duplicates()]
```

```python
raw_daily = raw_data.resample(rule="D").sum()
raw_daily = raw_daily[:-1]
```

```python
raw_daily.plot(figsize=(10,4))
```
